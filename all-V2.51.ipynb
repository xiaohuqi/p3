```python
def is_num(x):
    try:
        float(x)
        return True
    except:
        return False
```

```python
data=[]   
all_information = all_information.fillna("")
for one_line in all_information.values:
    data.append(one_line) 
```

```python
data[19638]
```

```python

strs_data = ['drug_class_7','drug_name','drug_concept_id','height','weight','chuyuan_disease','chuyuan_disease_id','gengzheng_disease',
             'gengzheng_disease_id','menzhen_disease','menzhen_disease_id','ruyuan_disease','ruyuan_disease_id','anesthesia_source_value','anesthesia_source_value',
            'operation_source_value','symptom_concept_id','symptom_name','note_text','measurement_datetime','examine_datetime','gengzheng_disease_datetime','de_start_datetime',
            'ruyuan_disease_datatime','chuyuan_disease_datetime','measurement_name','measurement_concept_id','examine_concept_id','examine_name'
            ,'examine_datetime','structure_text']
time_data=['gengzheng_disease_datetime','de_start_datetime',"ruyuan_disease_datatime","chuyuan_disease_datetime"]
date_condition=[]


for value in data:
    source_data = dict()
    for i in range(len(value)):
        try:
            if np.isnan(value[i]):
                source_data[key_i[i]] = value[i]
            else:
                source_data[key_i[i]] = value[i]
                if key_i[i] in strs_data:
                    if key_i[i] in time_data:
                        source_data[key_i[i]] = value[i].split("##")[0]
                    else:
                        try:
                            source_data[key_i[i]] = value[i].split("##")
                        except:
                            source_data[key_i[i]] = str(value[i]).split("##")
                else:
                    source_data[key_i[i]] = value[i]
        except:
            if key_i[i] in strs_data:
                if key_i[i] in time_data:
                    source_data[key_i[i]] = value[i].split("##")[0]
                else:
                    source_data[key_i[i]] = value[i].split("##")
            else:
                source_data[key_i[i]] = value[i]
    date_condition.append(source_data)
```

```python
date_condition[444]
```

```python
new_dict = dict()
for each in date_condition:
    new_dict[each['visit_occurrence_id']] = dict()
    new_dict[each['visit_occurrence_id']]['data'] = each
    new_dict[each['visit_occurrence_id']]['measurement'] = []
    new_dict[each['visit_occurrence_id']]['examine'] = []
    
measurement_data = pd.read_excel('measurement_data.xlsx')
measurement = []
measurement_data = measurement_data.fillna("")
for one_line in measurement_data.values:
    measurement.append(one_line) 
    
examine_data = pd.read_excel('procedure_examine_data.xlsx')
examine = []
examine_data = examine_data.fillna("")
for one_line in examine_data.values:
    examine.append(one_line)
        
for each in measurement:
    if each[0]:
        each_measurement_dict = dict()
        each_measurement_dict['concept_id'] = str(each[1])
        each_measurement_dict['name'] = each[2]
        if is_num(each[3]):
            each_measurement_dict['number_value'] = float(each[3])
        else:
            each_measurement_dict['other_value'] = each[3]
        each_measurement_dict['datetime'] = str(each[4])
        new_dict[each[0]]['measurement'].append(each_measurement_dict)
for each in examine:
    if each[0]:
        each_examine_dict = dict()
        each_examine_dict['concept_id'] = str(each[1])
        each_examine_dict['name'] = each[2]+each[3]+each[4]+each[5]
        if is_num(each[6]):
            each_examine_dict['number_value'] = float(each[6])
        else:
            each_examine_dict['other_value'] = each[6]
        each_examine_dict['datetime'] = str(each[7])
        new_dict[each[0]]['examine'].append(each_examine_dict)
new_data = []
for each in date_condition:
    new_dict[each['visit_occurrence_id']]['data']['measurement'] = new_dict[each['visit_occurrence_id']]['measurement']   
    new_dict[each['visit_occurrence_id']]['data']['examine'] = new_dict[each['visit_occurrence_id']]['examine']
    new_dict[each['visit_occurrence_id']]['data']['measurement_0'] = new_dict[each['visit_occurrence_id']]['measurement']   
    new_dict[each['visit_occurrence_id']]['data']['examine_0'] = new_dict[each['visit_occurrence_id']]['examine']
    new_data.append(new_dict[each['visit_occurrence_id']]['data'])
```

```python
index_condition_symptom1="all_information_v7e"
type_condition_symptom1="z_all_information_v7e"

information=[]
for item in new_data:
    action = {
                "_index": index_condition_symptom1,
                "_type": type_condition_symptom1,
    #           "_id":i,
                "_source": item
            }
    information.append(action)
#     print(i)
print(information[0])
```

```python
success=helpers.bulk(es,information)
print('Performed %d action'%(success[0]))
```
